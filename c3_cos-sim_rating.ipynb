```
# rating utility matrix function

def user_rating_matrix(df, val='rating', idx='user_id', col='item_id'):
  result = df.pivot_table(values=val, index=idx, columns=col)
  return result.fillna(result.mean())
```

```
# cosine similarity function for item_id

def cosine_sim(ratings_mat, item_id):
  return cosine_similarity(ratings_mat[item_id].values.reshape(1,-1), ratings_mat.T)[0]
```
